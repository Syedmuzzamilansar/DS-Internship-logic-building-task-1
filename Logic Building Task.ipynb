## Task 1: User Login Check
This program checks whether the username and password match the given credentials.


```python
# Task 1: User Login Check

correct_username = "admin"
correct_password = "1234"

username = input("Enter username: ")
password = input("Enter password: ")

if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## Task 2: Pass / Fail Analyzer

This program analyzes student marks and counts total pass and fail students.
A student passes if marks ≥ 50.


```python
# Task 2: Pass / Fail Analyzer (Upgraded Version)

marks = [45, 78, 90, 33, 60]

pass_students = [m for m in marks if m >= 50]
fail_students = [m for m in marks if m < 50]

print("Total Students:", len(marks))
print("Passed Students:", len(pass_students))
print("Failed Students:", len(fail_students))
```

## Task 3: Simple Data Cleaner

This program removes extra spaces and converts names to lowercase.


```python
# Task 3: Simple Data Cleaner (Upgraded Version)

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]

print("Cleaned Names:", cleaned_names)
```

## Task 4: Message Length Analyzer

This program calculates message length and flags messages longer than 10 characters.


```python
# Task 4: Message Length Analyzer (Upgraded Version)

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    status = "Long Message" if length > 10 else "Valid Length"

    print(f"Message: '{message}'")
    print(f"Length: {length}")
    print(f"Status: {status}")
    print("-" * 30)
```

## Task 5: Error Message Detector

This program counts the total number of ERROR entries in system logs.


```python
# Task 5: Error Message Detector (Upgraded Version)

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")

print("Total ERROR messages:", error_count)
```
